```python
# Sistem Pakar Diagnosa Penyakit (Rule-Based AI)

# Daftar gejala
gejala_list = [
    "Demam tinggi",
    "Batuk kering",
    "Batuk berdahak",
    "Pilek",
    "Hidung tersumbat",
    "Sakit tenggorokan",
    "Sesak napas",
    "Nyeri dada",
    "Badan lemas",
    "Sakit kepala"
]

# Input dari user
print("=== SISTEM PAKAR DIAGNOSA PENYAKIT ===")
print("Jawab dengan 'y' atau 'n'\n")

input_gejala = {}

for gejala in gejala_list:
    jawab = input(f"Apakah Anda mengalami {gejala}? (y/n): ").lower()
    input_gejala[gejala] = jawab == 'y'

# Rule-based diagnosis
def diagnosa(g):
    # Pneumonia
    if (g["Demam tinggi"] and g["Batuk berdahak"] and
        g["Sesak napas"] and g["Nyeri dada"] and g["Badan lemas"]):
        return "Pneumonia"

    # Bronkitis
    elif (g["Batuk berdahak"] and g["Nyeri dada"] and
          g["Sesak napas"] and g["Badan lemas"]):
        return "Bronkitis"

    # Influenza
    elif (g["Demam tinggi"] and (g["Batuk kering"] or g["Batuk berdahak"]) and
          g["Pilek"] and g["Sakit tenggorokan"] and g["Badan lemas"]):
        return "Influenza"

    # Asma
    elif (g["Sesak napas"] and g["Nyeri dada"]):
        return "Asma"

    else:
        return "Tidak terdeteksi / Perlu pemeriksaan lebih lanjut"

# Hasil diagnosadiagnosa
hasil = diagnosa(input_gejala)

print("\n=== HASIL DIAGNOSA ===")
print("Kemungkinan penyakit:"diagnosa
```

```python
# Import library
import numpy as np
from sklearn.linear_model import LinearRegression

# Data training sederhana (contoh)
# Format: [UTS, UAS]
X = np.array([
    [60, 65],
    [70, 75],
    [80, 85],
    [90, 95],
    [50, 55],
    [85, 80],
    [75, 70],
    [95, 90]
])

# Target nilai akhir (misalnya rata-rata + sedikit variasi)
y = np.array([62, 73, 83, 93, 52, 82, 72, 92])

# Buat model
model = LinearRegression()
model.fit(X, y)

# Input dari user
uts = float(input("Masukkan nilai UTS: "))
uas = float(input("Masukkan nilai UAS: "))

# Prediksi nilai akhir
prediksi = model.predict([[uts, uas]])[0]

# Fungsi klasifikasi nilai
def kategori_nilai(nilai):
    if nilai >= 90:
        return "A (Sangat Baik)"
    elif nilai >= 80:
        return "B (Baik)"
    elif nilai >= 70:
        return "C (Cukup)"
    else:
        return "D (Kurang)"

# Output
print("\n=== HASIL PREDIKSI ===")
print(f"Nilai akhir: {prediksi:.2f}")
print(f"Grade: {kategori_nilai(prediksi)}")
```
